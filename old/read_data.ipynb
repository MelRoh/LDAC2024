```python
from SPARQLWrapper import SPARQLWrapper, JSON
import pandas as pd
import json

# Function to query DBpedia
def query_dbpedia(sparql_query):
    sparql = SPARQLWrapper("http://dbpedia.org/sparql")
    sparql.setQuery(sparql_query)
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()
    
    return results["results"]["bindings"]
```

```python
import requests
import json
from bs4 import BeautifulSoup

# Wikipedia API endpoint
WIKIPEDIA_API_URL = "https://de.wikipedia.org/w/api.php"

# Function to query Wikipedia for "Schachte in Bochum"
def query_wikipedia(query):
    params = {
        "action": "query",
        "list": "search",
        "srsearch": query,
        "format": "json",
        "srlimit": 100
    }
    response = requests.get(WIKIPEDIA_API_URL, params=params)
    return response.json()

# Extract relevant information from the query result
def extract_information(data):
    shafts = []
    for item in data['query']['search']:
        shaft_info = {
            "title": item['title'],
            "snippet": item['snippet'],
            "pageid": item['pageid']
        }
        shafts.append(shaft_info)
    return shafts

# Function to get latitude and longitude from Zechenkarte
def get_lat_long(title):
    url = f"http://www.zechenkarte.de/wiki/index.php?title={title.replace(' ', '_')}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    lat_long = {"latitude": None, "longitude": None}
    geo_data = soup.find("span", class_="coordinates")
    if geo_data:
        coords = geo_data.text.split(';')
        if len(coords) == 2:
            lat_long['latitude'] = coords[0].strip()
            lat_long['longitude'] = coords[1].strip()
    
    return lat_long

# Main function

query = "Alle Schächte"
data = query_wikipedia(query)
shafts_info = extract_information(data)

for shaft in shafts_info:
    title = shaft['title']
    lat_long = get_lat_long(title)
    shaft.update(lat_long)

# Print the result as JSON
print(json.dumps(shafts_info, indent=4, ensure_ascii=False))


```

```python


# Function to convert to JSON-LD
def to_json_ld(shafts):
    json_ld = {
        "@context": {
            "@vocab": "http://schema.org/",
            "latitude": "geo:latitude",
            "longitude": "geo:longitude"
        },
        "@graph": []
    }
    for shaft in shafts:
        shaft_ld = {
            "@type": "Place",
            "name": shaft['title'],
            "description": shaft['snippet'],
            "url": f"https://de.wikipedia.org/?curid={shaft['pageid']}",
            "geo": {
                "@type": "GeoCoordinates",
                "latitude": shaft['latitude'],
                "longitude": shaft['longitude']
            }
        }
        json_ld["@graph"].append(shaft_ld)
    return json_ld

# Main function
query = "Schachte in Bochum"
data = query_wikipedia(query)
shafts_info = extract_information(data)

for shaft in shafts_info:
    title = shaft['title']
    lat_long = get_lat_long(title)
    shaft.update(lat_long)

json_ld = to_json_ld(shafts_info)

# Print the result as JSON-LD
print(json.dumps(json_ld, indent=4, ensure_ascii=False))


```

```python
import requests
from bs4 import BeautifulSoup

# URL of the page
url = "http://www.zechenkarte.de/wiki"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find all <script> tags containing leaflet data
scripts = soup.find_all('script')

# Iterate over each <script> tag to find leaflet data
for script in scripts:
    # Check if the script contains leaflet data
    if "var map = L.map" in script.text:
        # Extract the leaflet data
        leaflet_data = script.text
        break

# Extract the lines containing marker data
marker_lines = [line.strip() for line in leaflet_data.split('\n') if 'L.marker' in line]

# Iterate over marker lines to extract name, latitude, and longitude
for line in marker_lines:
    # Extract name
    name_start = line.find("'") + 1
    name_end = line.find("'", name_start)
    name = line[name_start:name_end]

    # Extract latitude and longitude
    lat_start = line.find('[') + 1
    lat_end = line.find(',', lat_start)
    latitude = line[lat_start:lat_end]

    lon_start = lat_end + 1
    lon_end = line.find(']', lon_start)
    longitude = line[lon_start:lon_end]

    print("Name:", name)
    print("Latitude:", latitude)
    print("Longitude:", longitude)
    print()

```

```python
import requests
from bs4 import BeautifulSoup
import json
import re

# URL of the main page to scrape
url = "http://www.zechenkarte.de/wiki/index.php?title=Kategorie:Schacht_in_Bochum"

# Send a GET request to the main webpage
response = requests.get(url)
response.raise_for_status()  # Check that the request was successful

# Parse the main webpage content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the list of shafts
shafts = []
for li in soup.select('.mw-category-group ul li'):
    title = li.find('a').text
    link = li.find('a')['href']
    shafts.append({"title": title, "link": "http://www.zechenkarte.de" + link})

# Function to convert DMS (Degrees, Minutes, Seconds) to Decimal Degrees
def dms_to_dd(degrees, minutes, seconds, direction):
    dd = float(degrees) + float(minutes)/60 + float(seconds)/(60*60)
    if direction in ['S', 'W']:
        dd *= -1
    return dd

# Function to get latitude and longitude from a shaft page
def get_shaft_details(shaft_url):
    response = requests.get(shaft_url)
    response.raise_for_status()
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract detailed information
    details = {"latitude": None, "longitude": None}
    geo_data = soup.find("a", class_="external text", href=re.compile("geohack"))
    print(geo_data)
    if geo_data:
        params = geo_data['href'].split('params=')[1].split('_')
        lat_deg, lat_min, lat_sec, lat_dir = params[0], params[1], params[2], params[3]
        lon_deg, lon_min, lon_sec, lon_dir = params[5], params[6], params[7], params[8]
        
        latitude = dms_to_dd(lat_deg, lat_min, lat_sec, lat_dir)
        longitude = dms_to_dd(lon_deg, lon_min, lon_sec, lon_dir)
        
        details['latitude'] = latitude
        details['longitude'] = longitude
    
    return details

# Update the main script to fetch details for each shaft
for shaft in shafts:
    details = get_shaft_details(shaft["link"])
    shaft.update(details)

# Print the results as JSON
print(json.dumps(shafts, indent=4, ensure_ascii=False))
```
